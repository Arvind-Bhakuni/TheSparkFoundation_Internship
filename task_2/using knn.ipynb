#### Building Model

```python
X = df.drop(['Species', 'Species_type'], axis=1)
y = df['Species_type']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=8)
```

```python
X.shape
```

```python
y.shape
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train, y_train)
```

```python
y_hat_train = knn.predict(X_train)
y_hat_test = knn.predict(X_test)
```

```python
accuracy_score(y_hat_train, y_train)
```

```python
accuracy_score(y_hat_test, y_test)
```

```python
def select_neighbors():
    knn_train_acc = []
    knn_test_acc = []
    for i in range(1, 11):
        knn = KNeighborsClassifier(n_neighbors=i)
        knn.fit(X_train, y_train)
        y_train_predict = knn.predict(X_train)
        y_test_predict = knn.predict(X_test)
        knn_train_acc.append(accuracy_score(y_train_predict, y_train))
        knn_test_acc.append(accuracy_score(y_test_predict, y_test))
    return knn_train_acc, knn_test_acc
```

```python
knn_train, knn_test = select_neighbors()
```

```python
x = np.linspace(1, 10, 10)

plt.figure(figsize=(10, 5))
plt.plot(x, knn_test, color='green', label='test_line')
plt.plot(x, knn_train, color='blue', label='train_line')
plt.title('Number of Neighbors vs Accuracy')
plt.xticks(np.linspace(1, 10, 10))
plt.grid()
plt.legend()
plt.show()
```

#### We will choose neighbors as n=7 since it gives highest accuracy on both testing and training data

```python
train_acc = round(knn_train[6], 3)
test_acc = round(knn_test[6], 3)
```

```python
print(f"Accuracy Score on Training data is {train_acc}")
print(f"Accuracy Score on Testing data is {test_acc}")
```
